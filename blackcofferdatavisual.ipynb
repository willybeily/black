```python
import json 
import csv
```

```python
with open('data.json',encoding="utf8") as json_file: 
    data = json.load(json_file) 
```

```python
data_file = open('data_file.csv', 'w',encoding="utf8") 
```

```python
csv_writer = csv.writer(data_file)
```

```python
count = 0
  
for inf in data: 
    if count == 0: 
  
        # Writing headers of CSV file 
        header = inf.keys() 
        csv_writer.writerow(header) 
        count += 1
  
    # Writing data of CSV file 
    csv_writer.writerow(inf.values()) 
  
data_file.close() 
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
df.dropna(axis=0,inplace=True)
df['sector']= le.fit_transform(df['sector']) 
df['topic']= le.fit_transform(df['topic']) 
df['insight']=le.fit_transform(df['insight'])
df['region']=le.fit_transform(df['region'])
df['added']=le.fit_transform(df['added'])
df['pestle']=le.fit_transform(df['pestle'])
df['source']=le.fit_transform(df['source'])
df['country']=le.fit_transform(df['country'])
```
